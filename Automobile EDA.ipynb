## a. Working on NORMALIZED - LOSSES

```python
cars["normalized-losses"]
```

## Replace function

```python
cars["normalized-losses"].replace("?",np.nan,inplace = True)  # replace old value to new value permanently
```

```python
cars.head()
```

<b> replace function --- > replaces the old value into new value

Syntax : replace(old value,new value)  ---- > temporary change

replace( old value, new value, inplace = True ) ---- > permanent change

## Data type change

```python
cars["normalized-losses"] = cars["normalized-losses"].astype("float") # changing datatype permanently
```

```python
cars.info()
```

<b> the dtype of normalized - losses as been changed from object to float64

```python
cars["normalized-losses"].fillna(cars["normalized-losses"].mean(),inplace = True)   
# Fills the na value to mean of all value present in that column
```

# 2. HANDLING CATEGORICAL DATA :
<b>(string type object / any non numerical categorical data)


Machine understands only numeric type data, where as Machine doesnot understand non - numeric data

<b> To convert categorical data to numerical data:

There are 2 types of method to change categorical data to numerical data

<b> 1. one hot encoding            2. label encoding

## a. ONE HOT ENCODING METHOD

First seperate the Numerical Data and categorical data

### Categorial data

```python
cars_cat = cars.select_dtypes(object)
```

```python
cars_cat.info()
```

```python
cars_cat
```

```python
cars_cat["fuel-type"].value_counts()
```

```python
pd.get_dummies(cars_cat["fuel-type"])
```

```python
pd.get_dummies(cars_cat["make"])
```

<b> if we observe it is showing only 1 and 0 value for alfa-romero and audio. But for the rest of other columns its not showing any value, which is not representable. So to make representable instead of using Hot encoding method we need to use label encoding method

## b. LABEL ENCODING METHOD

<b> To import Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()  # to create an object of class LabelEncoder
```

```python
le.fit_transform(cars_cat["fuel-type"]) #gas = 1, diesel = 0
```

```python
le.fit_transform(cars_cat["make"])
```

```python
for col in cars_cat:
    le = LabelEncoder()
    cars_cat[col] = le.fit_transform(cars_cat[col])
```
